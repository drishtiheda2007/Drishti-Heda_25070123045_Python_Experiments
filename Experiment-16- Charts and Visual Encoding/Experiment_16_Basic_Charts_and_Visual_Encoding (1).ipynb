```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data={
    'Days':['Mon','Tue','Wed','Thru','Fri'],
    'Study_Hours':[2,3,2.5,4,3.5],
    'Marks':[50,60,55,70,65],
    'Attendance':[60,70,65,80,75],
    'Sleep_Hours':[6,7,6,8,7],
    'Assignment_Completed':[1,2,1,3,2]
}
df = pd.DataFrame(data)
df
```

```python
#1. Line Chart(Trend)
plt.plot(df['Days'], df['Study_Hours'], marker='o')
plt.show()
```

```python
plt.plot(df['Days'], df['Marks'], marker='s')
plt.title("Study hours Trend")
plt.xlabel("Days")
plt.ylabel("Study hours")
plt.show()
```

```python
plt.plot(df['Days'], df['Marks'], marker='s')
plt.title("Study hours Trend")
plt.xlabel("Days")
plt.ylabel("Study hours")
plt.show()
```

```python
#Advance line chart
plt.figure(figsize=(10,6))

plt.plot(df['Days'],df['Study_Hours'],
         marker='*',label='Study Hours', color='green')

plt.plot(df['Days'],df['Marks'],
         marker='s',color='aqua',label='Marks')

plt.title('Study Hours and Marks Trend')
plt.xlabel('Days')
plt.ylabel('Value')
plt.legend()
```

```python
#2. Bar Chart (comparison)
plt.bar(df['Days'], df['Marks'], color='pink')

plt.title('Marks Comparison')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.show()
```

```python
plt.bar(df['Study_Hours'], df['Marks'], color='lavender')

plt.title('Marks Comparison')
plt.xlabel('Study_Hours')
plt.ylabel('Marks')
plt.show()
```

```python
#Advanced bar chart
bars = plt.bar(df['Days'], df['Marks'], color='crimson')
for bar in bars:
  y = bar.get_height()
  plt.text(bar.get_x() + bar.get_width() / 2, y, str(y), ha='center', va='bottom')

plt.title('Marks Comparison')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.grid(axis='y')
plt.show() #adds horizontal grid
```

```python
#3. Histogram (Distribution)
#Shows how values are spread

data = { 'Marks': [50, 55,60,63,87,90,67,45,53,83,87,57,80,65,69]
}
df = pd.DataFrame(data)
plt.hist(df['Marks'], bins=5, edgecolor='black', alpha = 0.60) #alpha- shade of color, 0= white, bins- division
plt.title('Marks Distribution')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt
```

```python
#4. Scatter plot (Relationship)
import matplotlib.pyplot as plt
data={
    'Days':['Mon','Tue','Wed','Thru','Fri'],
    'Study_Hours':[2,3,2.5,4,3.5],
    'Marks':[50,60,55,70,65],
    'Attendance':[60,70,65,80,75],
    'Sleep_Hours':[6,7,6,8,7],
    'Assignment_Completed':[1,2,1,3,2]
}
df = pd.DataFrame(data)

plt.scatter(df['Study_Hours'], df['Marks'])

plt.title('Study Hours vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.show()
```

```python
#Conditional Scatter Plot
#Example Category
df['Result'] = ['Fail', 'Pass', 'Pass', 'Fail', 'Pass']
color = ['red' if r == 'Fail' else 'green' for r in df['Result']]
plt.scatter(df['Study_Hours'], df['Marks'], c=color)

plt.title("Study Hours vs Marks (Result-wise)")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

```python
#Dataset-2
data = {
    'Day': ['Mon','Tue','Wed','Thu','Fri','Sat','Sun','Mon','Tue','Wed'],
    'Region': ['North','South','East','West','North','South','East','West','North',
               'South'],
    'Sales': [200,220,250,270,300,350,400,280,260,290],
    'Profit': [20,25,30,35,50,60,80,40,38,45],
    'Customers': [30,35,40,45,50,65,80,55,48,52],
    'Category': ['Food','Clothing','Electronics','Food','Clothing','Electronics',
                 'Food','Clothing','Electronics','Food']
}

df = pd.DataFrame(data)
print(df)
```

```python
#1. Line Chart(Trend)
plt.plot(df['Day'], df['Sales'], marker='*')
plt.title("Day vs Sales")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.show()
```

```python
#Advance line chart
plt.figure(figsize=(8,4))

plt.plot(df['Day'],df['Sales'],
         marker='*',label='Sales', color='orange')

plt.plot(df['Day'],df['Profit'],
         marker='s',color='cyan',label='Profit')

plt.title('Sales vs Profit Trend')
plt.xlabel('Days')
plt.ylabel('Value')
plt.legend()
```

```python
#2. Bar chart
plt.bar(df['Day'], df['Profit'], color='pink')

plt.title('Profit Comparison')
plt.xlabel('Day')
plt.ylabel('Profit')
plt.show()
```

```python
plt.bar(df['Sales'], df['Profit'], color='green')

plt.title('Profit Comparison')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

```python
#Advanced bar chart
bars = plt.bar(df['Day'], df['Profit'], color='orange')
for bar in bars:
  y = bar.get_height()
  plt.text(bar.get_x() + bar.get_width() / 2, y, str(y), ha='center', va='bottom')

plt.title('Profit Comparison')
plt.xlabel('Day')
plt.ylabel('Profit')
plt.grid(axis='y')
plt.show()
```
