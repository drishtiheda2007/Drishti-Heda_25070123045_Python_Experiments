```python
# Tuples in Python

mytuple=("Drishti","Jayvee","Tanmay")
print(mytuple[1])
```

```python
# Tuples in Python

mytuple=("Drishti","Jayvee","Tanmay")
mytuple[1]="Pallavi"
print(mytuple[1])
```

```python
# Append Items in Tuples

mytuple=("Drishti","Jayvee","Tanmay")
mytuple.append("Pallavi")
print(mytuple)
```

```python
# Add Tuple to a Tuple
mytuple=("Kina","Rory","Glimore")
mytuple2=("Gentle Reader",)
mytuple+=mytuple2
print(mytuple)
```

```python
list1 = (20,30,40,50,60,70)

print(list1[1:3])
print(list1[1:3:1])
print(list1[1:5:2])
print(list1[1:4:3])
print(list1[6:4:-1])
```

```python
list1 = (20,30,40,50,60,10,70)

print(list1[1:3])
print(list1[1:3:1])
print(list1[1:5:2])
print(list1[1:4:3])
print(list1[6:4:-2])
print(list1[-2:-4:-2])
```

```python
exam_result = ("Physics", 55, "C")

# Unpack the tuple elements into separate variables
subject, marks, grade = exam_result

# Display the subject, marks, and grade
print(f"Subject: {subject}")
print(f"Marks Obtained: {marks}")
print(f"Grade: {grade}")

# Add if condition to check for 75 marks
if marks >= 75:
    print("Student scored 75 or more marks.")
else:
    print("Student scored less than 75 marks.")
```
