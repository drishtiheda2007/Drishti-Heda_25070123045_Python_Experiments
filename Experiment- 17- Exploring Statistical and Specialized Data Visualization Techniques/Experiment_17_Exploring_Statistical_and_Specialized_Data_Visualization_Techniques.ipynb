```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
np.random.seed(1) #seed: to have same data for everyone
df= pd.DataFrame({
    'Category' : ['A', 'B', 'C', 'D', 'E'],
    'Values': [0,70,30,90,60],
    'Sales' : np.random.randint(100,500,5),
    'Profit' : np.random.randint(10,100,5)
})
print(df.head())
```

```python
#1. Area Plot
plt.figure()
plt.fill_between(df['Category'], df['Values'], color='pink', alpha = 0.40)
plt.title('Area Plot')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

```python
sns.set_style("whitegrid")  #apply seaborn style
plt.figure(figsize=(10,6))
plt.fill_between(df['Category'], df['Sales'], color='red', alpha = 0.70, label='Sales')

plt.fill_between(df['Category'], df['Profit'], color='lightblue', alpha = 0.65, label='Profit')

plt.title('Sales and Profit Area Plot')
plt.xlabel('Category')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
#2. Pie Chart
plt.figure()
plt.pie(df['Values'], labels =df['Category'], autopct= '%1.2f%%') #autopct- 0.2= 2 decimals, 0.1= 1 decimal
plt.title('Pie Chart')
plt.show()
```

```python
plt.figure()
explode = [0,0.3,0,0,0]
plt.pie(df['Values'], labels =df['Category'], autopct= '%1.2f%%', explode = explode) #autopct- 0.2= 2 decimals, 0.1= 1 decimal
plt.title('Pie Chart')
plt.show()
```

```python
#3. Donut chart
plt.figure()
plt.pie(df['Values'], labels =df['Category'], autopct= '%1.2f%%')
centre_circle = plt.Circle((0,0), 0.60, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("Donut Chart")
plt.show()
```

```python
#4. Boxplot (Outliers Detection)
plt.figure()
sns.boxplot(x=df['Values'])
plt.title("Boxplot of Sales")
plt.show()
```

```python
#5. Heatmap (Correlation)
plt.figure()
corr = df[['Sales', 'Profit']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
corr = df[['Values','Sales','Profit']].corr()

plt.figure(figsize=(5,4))

sns.heatmap(corr, annot=True)

plt.title("Correlation Heatmap")

plt.show()
```

```python
#Bubble Plot (3 Variable Chart)
import matplotlib.pyplot as plt

plt.scatter(df['Sales'], df['Profit'],
            s=df['Values']*8)   # size of bubbles

plt.title("Bubble Plot")
plt.xlabel("Sales")
plt.ylabel("Profit")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,7))

sns.scatterplot(x='Sales', y='Profit',
                size='Values',        # bubble size
                hue='Values',         # color
                data=df,
                sizes=(70, 250),      # min-max size
                palette='viridis')

plt.title("Bubble Plot using Seaborn")

plt.show()
```

```python
#2. Second Dataset
np.random.seed(7)
n = 100
df = pd.DataFrame({
    'Customer_ID': range(1, n+1),
    'Age': np.random.randint(20, 60, n),
    'Income': np.random.randint(20000, 100000, n),
    'Loan_Amount': np.random.randint(5000, 50000, n),
    'Credit_Score': np.random.randint(300, 900, n)
})

df['Loan_Status'] = np.where(df['Credit_Score'] > 600, 'Approved', 'Rejected')

print(df.head())
```

```python
#Boxplot (Outlier Detection – Loan Amount)
import matplotlib.pyplot as plt

plt.boxplot(df['Loan_Amount'])

plt.title("Loan Amount Distribution (Boxplot)")
plt.ylabel("Loan Amount")

plt.show()
```

```python
#3. Histogram (Credit Score Distribution)
plt.hist(df['Credit_Score'], bins=6, color='green', edgecolor='black')

plt.title("Credit Score Distribution")
plt.xlabel("Credit Score")
plt.ylabel("Frequency")

plt.show()
```

```python
#Bubble Plot (Advanced Scatter)
plt.scatter(df['Income'], df['Loan_Amount'],
            s=df['Credit_Score']/2,
            c=df['Credit_Score'],
            cmap='viridis',
            alpha=0.6)

plt.colorbar(label='Credit Score')

plt.title("Bubble Plot (Income vs Loan)")
plt.xlabel("Income")
plt.ylabel("Loan Amount")

plt.show()
```

```python
#6. Correlation Heatmap (Very Important)
import seaborn as sns

corr = df[['Age','Income','Loan_Amount','Credit_Score']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title("Correlation Matrix")

plt.show()
```
