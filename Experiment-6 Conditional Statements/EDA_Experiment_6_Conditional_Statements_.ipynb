## 5. Get input from user for 5 subjects and find average. Caluclate grade using if-elif-else.


```python
marks1 = int(input("Enter marks 1: "))
marks2 = int(input("Enter marks 2: "))
marks3 = int(input("Enter marks 3: "))
marks4 = int(input("Enter marks 4: "))
marks5 = int(input("Enter marks 5: "))

avg = (marks1 + marks2 + marks3 + marks4 + marks5) / 5
print("Average =", avg)

if avg >= 90:
    print("Grade A")
elif avg >= 75:
    print("Grade B")
elif avg >= 60:
    print("Grade C")
elif avg >= 40:
    print("Grade D")
else:
    print("Fail")
```

## 7. Python Program to check if a Date id valid and print the Incremented Date if it is valid.

```python
# take input
x = input("Enter day/month/year: ").split("/")

d = int(x[0])
m = int(x[1])
y = int(x[2])

# check leap year
leap = False
if (y % 4 == 0 and y % 100 != 0) or (y % 400 == 0):
    leap = True

# find max days in month
if m in [1,3,5,7,8,10,12]:
    max_d = 31
elif m in [4,6,9,11]:
    max_d = 30
elif m == 2:
    if leap:
        max_d = 29
    else:
        max_d = 28
else:
    max_d = 0

# check valid date
if d >= 1 and d <= max_d:
    print("Valid date")

    # find next date
    d = d + 1

    if d > max_d:
        d = 1
        m = m + 1

        if m > 12:
            m = 1
            y = y + 1

    # print next date
    print("Next date:", d, "/", m, "/", y, sep="")

else:
    print("Invalid date")
```

## 9.Tax calculation on the basis of income

Write a python program to calculate income tax based on annual income:

Up to 2,50,000: No tax

2,50,001 - 5,00,000: 5%

Above 10,00,000: 20%

```python
income = float(input("Enter your annual income (in INR): "))
tax = 0.0

if income <= 300000:
    tax = 0
elif income <= 600000:
    tax = (income - 300000) * 0.05
elif income <= 900000:
    tax = 15000 + (income - 600000) * 0.10  # 5% on 3L (15000) + 10% on next slab
elif income <= 1200000:
    tax = 15000 + 30000 + (income - 900000) * 0.15 # 5% on 3L (15000) + 10% on 3L (30000) + 15% on next slab
elif income <= 1500000:
    tax = 15000 + 30000 + 45000 + (income - 1200000) * 0.20 # 5% on 3L (15000) + 10% on 3L (30000) + 15% on 3L (45000) + 20% on next slab
else:
    tax = 15000 + 30000 + 45000 + 60000 + (income - 1500000) * 0.30 # Sum of taxes from previous slabs + 30% on remaining

print(f"Your Income is ₹{income:,.2f}, the calculated tax is ₹{tax:,.2f}.")
```
