```python
# Problem Statement - 1

marks = [78, 85, 90, 66, 88]

print("Maximum Marks:", max(marks))
print("Minimum Marks:", min(marks))
print("Total Marks:", sum(marks))
print("Number of Students:", len(marks))
average = sum(marks) / len(marks)
print("Average Marks:", average)

# Sort the marks in ascending order:

marks.sort()
print("Marks in Ascending Order:", marks)



```

```python
#Problem Statement - 2

student_marks = [80 , 88 , 95, 99 , 96, 75, 94, 90, 78, 69]
print("Highest Mark - ", max(student_marks))
print("Lowest Mark - ", min(student_marks))
print("Average Mark - ", sum(student_marks)/len(student_marks))
student_marks.sort()
print("Ascending Sorted Marks:", student_marks)
student_marks.sort(reverse = True)
print("Descending Sorted Marks:", student_marks)
```

```python
'''3. Grocery Shopping List
A user maintains a grocery lsit for montly shopping.
Tasks:
1. Create a grocery List
2. Add new items to the list
3. Remove purchased items
4. Display the final shopping list
Concepts Used:  append(),  remove() '''


grocery_list = ["Rice", "Maggi Masala", "Yogurt", "Chocolate"]
print("Initial Grocery List:", grocery_list)

grocery_list.append("Vegetables")
grocery_list.append("Fruits")
print("After Adding Items:", grocery_list)

grocery_list.remove("Rice")
grocery_list.remove("Maggi Masala")
print("After Removing Purchased Items:", grocery_list)

print("Final Shopping List:", grocery_list)
```

```python
'''4.A person stores phone contacts in a list.
Tasks:
1. Create a list of contacts
2. Add a new contact
3. Delete an existing contact
4. Display contacts alphabetically
Concepts Used: append(), remove(), sort()'''


contacts = ["RJ", "Cory", "Jalin", "Kina"]
print("Initial Contacts List:", contacts)

contacts.append("Cory")
print("After Adding a Contact:", contacts)

contacts.remove("Kina")
print("After Deleting a Contact:", contacts)

contacts.sort()
print("Contacts in Alphabetical Order:", contacts)
```

```python
# Problem Statement - 5 - Temperature Analysis

temp = [ 34, 45, 38, 37, 32 , 29, 30 , 28, 31 , 30]
print("The first 5 days Temperature are: ", temp[0:5], "\n")
print("The last 5 days Temperature are: ", temp[-5:], "\n")
print("The highest Temperature: ", max(temp), "\n")
print("The lowest Temperature: ", min(temp), "\n")
print("The average Temperature: ", sum(temp)/len(temp), "\n")
```
