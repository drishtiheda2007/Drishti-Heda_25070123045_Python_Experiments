###Title: Data Normalization and turning Categorical Variables into Quantitative Variables in Python


```python
#Dataset
import pandas as pd
import numpy as np

data = {
    'Product':['Laptop','Mobile','Tablet','Headphones','Smartwatch','Camera'],
    'Price':[55000,20000,30000,2000,8000,45000],
    'Units_Sold':[25,60,35,80,50,15],
    'Discount':[10,15,12,20,18,8]
}

df=pd.DataFrame(data)
df
```

```python
#1. Min-Max Normalization
#Min-Max Normalization for Price

df['Price_MinMax']= (df['Price'] - df['Price'].min()) / (df['Price'].max() - df['Price'].min())

print("\nMin-Max Normalized Price")
(df[['Product','Price','Price_MinMax']])
```

```python
#1. Min-Max Normalization
#Min-Max Normalization for Price

df['Price_MinMax']= (df['Price'] - df['Price'].min()) / (df['Price'].max() - df['Price'].min())

print("\nMin-Max Normalized Price")
(df[['Product','Price','Price_MinMax']])
```

```python
#2. Z-Score Normalization
df['Units_Zscore'] = (df['Units_Sold'] - df['Units_Sold'].mean()) / df['Units_Sold'].std()
df
```

```python
#3.Decimal Scaling

df['Price_decimal'] = df['Price'] / 100000 #(10^5)

df
```

```python
#Normalize Multiple Columns
cols = ['Price','Units_Sold','Discount']

df_norm = (df[cols] - df[cols].min()) / (df[cols].max() - df[cols].min())
df_norm
```

#Part B- Turning Categorical Variables into Quantitative Variables in Python

```python
data = {
    'Order ID': [101, 102, 103, 104, 105, 106],
    'Customer_Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    "Payment_Method": ['Credit Card', 'PayPal', 'Credit Card', 'PayPal', 'Credit Card', 'PayPal'],
    'Product-category': ['Electronics', 'Clothing', 'Electronics', 'Clothing', 'Electronics', 'Clothing'],
    'City':['New York', 'Los Angeles', 'Chicago', 'San Francisco', 'Miami', 'Seattle'],
    'Order_value': [1500, 800, 2000, 1667, 12788, 7756]
}

df1=pd.DataFrame(data)
df1
```

```python
#Method 1: Label Encoding

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df1['Gender_Label'] = le.fit_transform(df1['Customer_Gender'])

df1
```

```python
#Method 2: One Hot Encoding

df_encoded = pd.get_dummies(df1, columns=['Payment_Method'])

df_encoded
```

```python
#Method 3: Dummy Encoding (Drop First)
df_dummy = pd.get_dummies(df1, columns=['Payment_Method'], drop_first=True)

df_dummy
```
