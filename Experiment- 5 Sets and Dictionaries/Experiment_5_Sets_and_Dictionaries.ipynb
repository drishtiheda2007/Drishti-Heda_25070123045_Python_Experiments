```python
'''Problem Statement 1: Unique Event Participants
A college event has a list of students who registered multiple times by mistake.
Write a program  using sets to store participants Names and display only unique participants.'''


# Problem Statement 1: Unique Event Participants

participants = ["Siya", "Diya", "Naino", "Divya", "Mahi", "Pari", "Purvi", "Manan", "Naino", "Mahi", "Siya"]

unique_participants = set(participants)

print("Unique Event Participants:")
for name in unique_participants:
    print(name)


```

```python
'''Problem Statement 2: Three students choode elective subjects. Find the common subjects chosen by all stuents using set oprations
Concept Used: Set intersection (&)'''

# Problem Statement 2: Common Elective Subjects

# Subjects chosen by three students
jayvee = {"EDA", "AMT", "PES", "DCLD"}
tanmay = {"ECSF", "FEEE", "AMT", "DCLD"}
jahnvi = {"AMT", "ECSF", "DCLD", "CALCULUS"}

# Find common subject
common_subjects = jayvee & tanmay & jahnvi

# Display result
print("Common subjects chosen by all students:")
for subject in common_subjects:
    print(subject)
```

```python
 '''Problem Statement 3: One set contains students in the Cricket Club, another in the Football Club.
Find students who are: 1. In both clubs 2. In only one club
Concepts Used: Intersection, Symmetric difference'''

# Problem Statement 3: Cricket Club and Football Club

# Students in each club
cricket_club = {"Neel", "Kanav", "Jignesh", "Dhruv", "Angad","Kunal"}
football_club = {"Angad", "Neel", "Dhruv", "Ojas", "Kunal"}

# 1. Students in both clubs (Intersection)
both_clubs = cricket_club & football_club

# 2. Students in only one club (Symmetric Difference)
only_one_club = cricket_club ^ football_club

# Display results
print("Students in both Cricket and Football clubs:")
for student in both_clubs:
    print(student)

print("\nStudents in only one club:")
for student in only_one_club:
    print(student)
```

```python
'''Problem Statement 4: Absent Students
Given a set of all students and a set of present  students, find the absent students.
Concept Used: Set Difference (-)'''

# Problem Statement 4: Absent Students

# Set of all students
all_students = {"Tanmay", "Jayvee", "Falak", "Hiranya", "Drishti", "Jahnvi", "Tannishtha", "Shreya", "Gunjan", "Jui"}

# Set of present students
present_students = {"Drishti", "Jayvee", "Jahnvi", "Tannishtha"}

# Find absent students using set difference
absent_students = all_students - present_students

# Display result
print("Absent Students:")
for student in absent_students:
    print(student)
```

```python
'''Problem Stateemnt: 5:  Remove Invalid Enteries.
A set contains vaild course codes.
Remove a discontinued course code from the set.
Concept Used: remove() / discard()'''


# Using discard() (no error if element not found)

course_codes = {"CS101", "AI202", "ML303", "MECH404"}
course_codes.discard("ML303")

print("Updated Course Codes:")
print(course_codes)


```
